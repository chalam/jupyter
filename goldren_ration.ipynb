```python
#golden ratio 
# a+b is to a as a is to b
#fixed-point of this function is this: 
# it is the place where y=xy=x (or f(x)=xf(x)=x )

# Plot y=(x+1)/x and y=x
x=np.linspace(0.5,3,1000)
y=map(lambda a: (a+1.0)/a, x)
y2=map(lambda a: a, x)
plt.plot(x, y)
plt.plot(x, y2, 'r')

# # Make graph square
fig = plt.gcf()
fig.set_size_inches(5,5)
```
